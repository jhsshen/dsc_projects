```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose the dataframe
properties_transposed = properties.transpose()
# Reset index
properties_transposed = properties_transposed.reset_index()
# Rename column names
new_columns = properties_transposed.iloc[0]
properties_transposed.columns = new_columns
# Remove the first row same as column names
properties_transposed = properties_transposed[1:]
properties_transposed.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename first two column names
properties_transposed.rename(columns = {'Unnamed: 0': 'borough', pd.NaT: 'geo_code'}, inplace = True)
properties_transposed
```

```python
# Found 'unnamed' in borough and 'NaN' in geo_code, remove rows with 'NaN' in geo_code
properties_transposed.dropna(subset='geo_code', inplace=True)
# Reset index
properties_transposed.reset_index(drop=True, inplace=True)
properties_transposed
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt wide table to long table
properties_transposed_long = properties_transposed.melt(id_vars=['borough', 'geo_code'], value_vars=properties_transposed.columns[2:], var_name='date', value_name='ave_price')
properties_transposed_long
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Check column structure and data types
properties_transposed_long.info()
# Convert 'ave_price' column to float type
properties_transposed_long['ave_price'] = properties_transposed_long['ave_price'].astype(float)
properties_transposed_long.dtypes
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Groupby 'borough' then 'year' and calculate annual average housing price of each borough
borough_year_mean = properties_long_32_boroughs_sort.groupby(['borough', 'year'])['ave_price'].agg(np.mean)
# Get the 32 boroughs list from the dataframe
boroughs_list = properties_long_32_boroughs_sort['borough'].unique().tolist()

# Fuction for price inceasing ratios (increased ratio by base price, and increasing speed by time passed)
def create_price_ratio(borough, start_year, end_year):
    '''
    Input: borough_name, start_year, end year.
    Output: borough name, start year, end year, ratio_1 (price difference/start_year_price), speed_2 (price difference/time difference)
    '''
    ratio_1 = (borough_year_mean[(borough, end_year)] - borough_year_mean[(borough, start_year)]) / borough_year_mean[(borough, start_year)]
    speed_2 = (borough_year_mean[(borough, end_year)] - borough_year_mean[(borough, start_year)]) / (end_year - start_year)
    return (borough, start_year, end_year, ratio_1, speed_2)

# Collect returned results in a list for a new dataframe, compare years of 2018 and 1998
ratio_list = []
for item in boroughs_list:
    result = create_price_ratio(item, 1998, 2018)
    ratio_list.append({'borough': result[0], 'start_year': result[1], 'end_year': result[2], 'ratio_1': result[3], 'speed_2': result[4]})

# Result dataframe
ratio_df = pd.DataFrame(ratio_list)

# Sort the result dataframe by price change relative ratio in descending order
print('From 1998 to 2018, average housing price has increased by ratio_1 times of the base price: ')
print(ratio_df.sort_values('ratio_1', ascending=False).head())

# Sort the result dataframe by price change speed in descending order
print('\nFrom 1998 to 2018, average housing price has averagely increased by speed_2 pounds per year: ')
print(ratio_df.sort_values('speed_2', ascending=False).head())
```
